```python
import pandas as pd

# بنحط المسار اللي أنت نسخته بين علامات تنصيص "" عشان بايثون يفهم إنه نص
file_path = "/content/drive/MyDrive/Hospital_Inpatient_Cost_Transparency_Beginning_2009_20240726.csv"

# السطر ده بيقرأ الملف ويخزنه في حاجة اسمها df
df = pd.read_csv(file_path)

# السطر ده هيعرضلك أول 5 سطور من الملف عشان تتأكد إنه فتح تمام
df.head()
```

```python
# حجم الداتا
print("Shape:", df.shape)

# معلومات الأعمدة
print(df.info())

# القيم الناقصة
print("\nMissing Values:")
print(df.isnull().sum())

# إحصائيات عامة
print("\nBasic Stats:")
print(df.describe())
```

```python
# تأكد من أسماء الأعمدة بالظبط
print(df.columns.tolist())
```

```python
# 1. نسخة من الداتا
df_clean = df.copy()

# 2. حذف الصفوف الناقصة
df_clean = df_clean.dropna(subset=['APR Severity of Illness Description',
                                    'APR Medical Surgical Code'])

# 3. حذف القيم الغريبة
df_clean = df_clean[df_clean['Mean Cost'] > 0]
df_clean = df_clean[df_clean['Mean Charge'] > 0]

# 4. إزالة الـ Duplicates
df_clean = df_clean.drop_duplicates()

# 5. تنظيف أسماء الأعمدة
df_clean.columns = df_clean.columns.str.strip().str.replace(' ', '_')

# 6. النتيجة
print("Before:", df.shape)
print("After:", df_clean.shape)
print("\nMissing Values after cleaning:")
print(df_clean.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. توزيع درجات الخطورة
plt.figure(figsize=(8,5))
df_clean['APR_Severity_of_Illness_Description'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Distribution of Severity of Illness')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. متوسط التكلفة حسب الخطورة
plt.figure(figsize=(8,5))
df_clean.groupby('APR_Severity_of_Illness_Description')['Mean_Cost'].mean().sort_values().plot(kind='bar', color='coral')
plt.title('Average Cost by Severity of Illness')
plt.xlabel('Severity')
plt.ylabel('Average Cost ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. متوسط التكلفة عبر السنين
plt.figure(figsize=(10,5))
df_clean.groupby('Year')['Mean_Cost'].mean().plot(kind='line', marker='o', color='green')
plt.title('Average Cost Over Years')
plt.xlabel('Year')
plt.ylabel('Average Cost ($)')
plt.tight_layout()
plt.show()
```

```python
# أغلى 10 تشخيصات
top_diagnosis = df_clean.groupby('APR_DRG_Description')['Mean_Cost'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_diagnosis.plot(kind='barh', color='steelblue')
plt.title('Top 10 Most Expensive Diagnoses')
plt.xlabel('Average Cost ($)')
plt.tight_layout()
plt.show()

# أغلى 10 مستشفيات
top_hospitals = df_clean.groupby('Facility_Name')['Mean_Cost'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_hospitals.plot(kind='barh', color='coral')
plt.title('Top 10 Most Expensive Hospitals')
plt.xlabel('Average Cost ($)')
plt.tight_layout()
plt.show()
```

```python
# طبي vs جراحي
plt.figure(figsize=(8,5))
df_clean.groupby('APR_Medical_Surgical_Description')['Mean_Cost'].mean().plot(kind='bar', color=['steelblue', 'coral', 'green'])
plt.title('Average Cost: Medical vs Surgical')
plt.xlabel('Type')
plt.ylabel('Average Cost ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# عدد المرضى حسب النوع
plt.figure(figsize=(8,5))
df_clean.groupby('APR_Medical_Surgical_Description')['Discharges'].sum().plot(kind='bar', color=['steelblue', 'coral', 'green'])
plt.title('Total Discharges: Medical vs Surgical')
plt.xlabel('Type')
plt.ylabel('Total Patients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')
df_clean.to_sql('hospital', conn, index=False)

# أعلى 5 مستشفيات في عدد المرضى
query = """
SELECT Facility_Name,
       SUM(Discharges) as Total_Patients,
       ROUND(AVG(Mean_Cost), 2) as Avg_Cost
FROM hospital
GROUP BY Facility_Name
ORDER BY Total_Patients DESC
LIMIT 5
"""
result = pd.read_sql(query, conn)
print(result)
```

```python
query2 = """
SELECT APR_DRG_Description,
       ROUND(AVG(Mean_Charge), 2) as Avg_Charge,
       ROUND(AVG(Mean_Cost), 2) as Avg_Cost,
       ROUND(AVG(Mean_Charge) - AVG(Mean_Cost), 2) as Difference
FROM hospital
GROUP BY APR_DRG_Description
ORDER BY Difference DESC
LIMIT 10
"""
result2 = pd.read_sql(query2, conn)
print(result2)
```
